```python
import numpy as np
from data.io import get_downloaded_dataset_ids, read_dataset
from data.config import Preprocess
from data.preprocessing.format import to_ndarray
from data.repo_local import get_downloaded_dataset_ids_by_task
```

```python
def diag_datasets(start_i=0):
    classification_dids = get_downloaded_dataset_ids_by_task('Supervised Classification')
    dataset_ids = get_downloaded_dataset_ids(Preprocess.RAW)
    num_datasets = len(dataset_ids)
    ns = []
    ds = []
    num_categoricals = []
    ks = []
    print('Reading {} in total'.format(num_datasets))
    for i in range(start_i, num_datasets):
        print(i + 1, end=' ')
        dataset_id = dataset_ids [i]
        d = read_dataset(dataset_id, preprocess=Preprocess.RAW)
        X, y, categorical = d['X'], d['y'], d['categorical']
        X = to_ndarray(X)
        n, d = X.shape
        num_categorical = sum(categorical)
        
        ns.append(n)
        ds.append(d)
        num_categoricals.append(num_categorical)
        
        if dataset_id in classification_dids:
            num_classes = len(np.unique(y))
            ks.append(num_classes)
        
    return ns, ds, num_categoricals, ks
```

```python
ns, ds, num_categoricals, ks = diag_datasets()
```
